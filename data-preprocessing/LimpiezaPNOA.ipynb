```python
import os

root_dir = r"G:\jorge\Datasets\Para_revisar y seleccionar"

subcarpetas = ["mask_selec", "ortho_selec"]

permitidas = {
    f.lower() for f in os.listdir(root_dir)
    if f.lower().endswith(".png") and os.path.isfile(os.path.join(root_dir, f))
}

print(f"Imágenes permitidas detectadas: {len(permitidas)}")

for sub in subcarpetas:
    sub_path = os.path.join(root_dir, sub)

    if not os.path.isdir(sub_path):
        print(f"Carpeta no encontrada: {sub_path}")
        continue

    for file in os.listdir(sub_path):
        if file.lower() not in permitidas:
            file_path = os.path.join(sub_path, file)
            try:
                os.remove(file_path)
                print(f"Eliminado (no permitido): {file_path}")
            except Exception as e:
                print(f"No se pudo eliminar {file_path}: {e}")
        else:
            print(f"Conservado: {file}")
```

```python
import os
import shutil
import random

random.seed(42)

root_dir = "G:\\jorge\\Datasets\\PNOA"
images_dir = os.path.join(root_dir, "chequeadas_ortho")
masks_dir = os.path.join(root_dir, "chequeadas_mask")

images = sorted(os.listdir(images_dir))
masks = sorted(os.listdir(masks_dir))

assert len(images) == len(masks), "Número de imágenes y máscaras no coinciden"

data = list(zip(images, masks))
random.shuffle(data)

n_total = len(data)
n_train = int(0.7 * n_total)
n_val = int(0.15 * n_total)

splits = {
    'train': data[:n_train],
    'val': data[n_train:n_train+n_val],
    'test': data[n_train+n_val:]
}

for split in ['train', 'val', 'test']:
    img_out = os.path.join(root_dir, split, "images")
    msk_out = os.path.join(root_dir, split, "masks")
    os.makedirs(img_out, exist_ok=True)
    os.makedirs(msk_out, exist_ok=True)

    for img_name, msk_name in splits[split]:
        shutil.copy(os.path.join(images_dir, img_name), os.path.join(img_out, img_name))
        shutil.copy(os.path.join(masks_dir, msk_name), os.path.join(msk_out, msk_name))

print("División completada.")
```
